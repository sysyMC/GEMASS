This script allow to get the publications associated to a pubmed id. Exemple from Emmanuel's dataset for pubmedCommons.

```python
import requests
import pandas as pd
from tqdm import tqdm
```

# scrape PMC papers

```python
dfPMC=pd.read_csv('../../Base de données/PubMed Commons_Base publications.csv',sep=';')
```

```python
dfPMC
```

```python
pmId=dfPMC['url']
```

```python
openalexId=[]
journalOAid=[]
authorsOAids=[]
baseURL='https://api.openalex.org/works?filter=ids.pmid:'
for pmid in tqdm(pmId):
    apiurl=baseURL+pmid
    cc=requests.get(apiurl).json()
    try:
        openalexId.append(cc['results'][0]['id'].split('/')[-1])
    except:
        openalexId.append('NONE')
    try:
        journalOAid.append(cc['results'][0]['host_venue']['id'].split('/')[-1])
    except:
        journalOAid.append('NONE')
    try:
        authorsOAids.append([i['author']['id'].split('/')[-1] for i in cc['results'][0]['authorships']])
    except:
        authorsOAids.append('NONE')
```

```python
dfPMC['OAid']=openalexId
dfPMC['journalOAid']=journalOAid
dfPMC['authors']=authorsOAids
```

```python
dfPMC.to_csv('PMC_OpenAlex.csv')
```
