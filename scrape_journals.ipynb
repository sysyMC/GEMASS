This script allow to get all the journals (venues) in openAlex and associate these last to the WOS disciplines. 

```python
import requests
import pandas as pd
from tqdm import tqdm
```

# JOURNALS

```python
venuesURL='https://api.openalex.org/venues?&mailto=floriana.gargiulo@cnrs.fr&per-page=100&cursor='
```

```python
cursor='*'
cc=requests.get(venuesURL+cursor).json()
```

```python
cc['results']
```

```python
journals=[]
cursor='*'
t=1
i=0
while t!=0:
    print(i*100)
    cc=requests.get(venuesURL+cursor).json()
    cursor=cc['meta']['next_cursor']
    journals=journals+cc['results']
    t=len(cc['results'])
    i=i+1
    
```

```python
oaID=[]
magId=[]
issn=[]
name=[]
nPaper=[]
nCitations=[]
openAccess=[]
for i in journals:
    oaID.append(i['id'])
    try:
        magId.append(i['ids']['mag'])
    except:
        magId.append('NONE')
    
    try:
        issn.append(i['issn'])
    except:
        issn.append('NONE')
    
    try:
        name.append(i['display_name'])
    except:
        name.append('NONE')
    
    try:
        nPaper.append(i['works_count'])
    except:
        nPaper.append('NONE')
        
    try:
        nCitations.append(i['cited_by_count'])
    except:
        nCitations.append('NONE')
        
    try:
        openAccess.append(i['is_oa'])
    except:
        openAccess.append('NONE')
```

```python
df=pd.DataFrame()
df['id_openAlex']=oaID
df['id_mag']=magId
df['issn']=issn
df['name']=name
df['nPaper']=nPaper
df['nCitations']=nCitations
df['IsOpenAccess']=openAccess
```

```python
dfRed=df[df['name']!='NONE']
```

```python
dfRed.to_csv('openAlexJournals.csv',index=False)
```

```python
from ast import literal_eval
def convertToList(stringa):
    try:
        return literal_eval(stringa)
    except:
        return 'NONE'
    
df=pd.read_csv('openAlexJournals.csv')
df['issn']=df.issn.apply(lambda x: convertToList(x))
```

```python
df
```

```python
dfWOS=pd.read_csv('wos/wos-core_AHCI 2022-February-23.csv')
dfWOS2=pd.read_csv('wos/wos-core_ESCI 2022-February-23.csv')
dfWOS3=pd.read_csv('wos/wos-core_SCIE 2022-February-23.csv')
dfWOS4=pd.read_csv('wos/wos-core_SSCI 2022-February-23.csv')

dfWOS=dfWOS.append(dfWOS2)
dfWOS=dfWOS.append(dfWOS3)
dfWOS=dfWOS.append(dfWOS4)

dfWOS=dfWOS.drop_duplicates()
```

```python
dfWOSissn=dict(zip(dfWOS['ISSN'],dfWOS['Web of Science Categories']))
dfWOStitle=dict(zip(dfWOS['Journal title'],dfWOS['Web of Science Categories']))
```

```python
df['lowName']=df['name'].apply(lambda x:x.lower())
assocByTitle={}
for i in dfWOStitle.keys():
    thisDf=df[df['lowName']==i.lower()]
    if len(thisDf)>0:   
        assocByTitle[list(thisDf['id_openAlex'])[0]]=dfWOStitle[i]
            
```

```python
len(assocByTitle)
```

```python
assocByISSN={}
redDf=df[df['issn']!='NONE']
for i in tqdm(dfWOSissn.keys()):
    mask = redDf.issn.apply(lambda x: i in x)
    thisDf=redDf[mask]
    if len(thisDf)>0:   
        assocByISSN[list(thisDf['id_openAlex'])[0]]=dfWOSissn[i]
```

```python
len(assocByISSN)
```

```python
for i in set(assocByTitle.keys()) & set(assocByISSN.keys()):
    if assocByTitle[i]!=assocByISSN[i]:
        del assocByTitle[i]
        del assocByISSN[i]
```

```python
allkeys=list(set(assocByTitle.keys()) | set(assocByISSN.keys()))
```

```python
completeDict={}
for i in allkeys:
    try:
        completeDict[i]=assocByISSN[i]
    except:
        completeDict[i]=assocByTitle[i]
```

```python
df['isWOS']=[1 if i in completeDict.keys() else 0 for i in df['id_openAlex']]
```

```python
dfOA_WOS=df[df['isWOS']==1]
```

```python
dfOA_WOS['WOS_disciplines']=dfOA_WOS['id_openAlex'].apply(lambda x:completeDict[x])
```

```python
dfOA_WOS.to_csv('OpenAlex_WOSdisciplines.csv',index=False)
```
